### Step 1 -  Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2 -Import Dataset

```python
data= pd.read_csv("dirty_data.csv")
data.head(5)    # get 5 rows
```

### Step 3 - Perform EDA

```python
# get columns names present in dataset
data.columns
```

```python
# obtaining basic statistical Summary off the data set.
data.describe()
```

```python
# get information about dataset
data.info()
```

```python
# get Count Of Null values In Dataset
data.isnull().sum()
```

```python
data['nearest_warehouse'].value_counts()
```

```python
data['season'].value_counts()
```

```python
# Drop unwanted Columns 
data.drop(columns=['latest_customer_review' , 'customer_long', 'distance_to_nearest_warehouse'])
```

```python
# convert Ware house into same form
data['nearest_warehouse'] = data['nearest_warehouse'].str.capitalize()
data
```

```python
# Convert summer data into same format
data['season'] = data['season'].str.capitalize()
data
```

```python
data['is_happy_customer']=data['is_happy_customer'].astype(int)
data['is_expedited_delivery']=data['is_expedited_delivery'].astype(int)
data
```

```python
# check in the data don't have any duplicate orders
duplicate_orders = data[data.duplicated(subset=['order_id', 'customer_id'], keep=False)]
duplicate_orders
```

```python
# get value counts and find out how many warehouses in the data
data['nearest_warehouse'].value_counts()
```

###  in below code we get only Nickolson Warehouse

```python
# Only Find Out Nickolson information 
Nickolson_column=data[data['nearest_warehouse']=='Nickolson']
Nickolson_column
```

```python
# filter nickolson data 
Nickolson_data=Nickolson_column.drop(columns=['order_id','date','shopping_cart','delivery_charges','customer_lat','customer_long','coupon_discount','distance_to_nearest_warehouse','latest_customer_review','latest_customer_review'])
```

### In this Data we get true or false data into 1 or 0 values

```python
Nickolson_data['is_happy_customer']=Nickolson_data['is_happy_customer'].astype(int)
Nickolson_data['is_expedited_delivery']=Nickolson_data['is_expedited_delivery'].astype(int)
Nickolson_data
```

```python
# chech happy customers in spring season
customer_Nickolson_data = Nickolson_data[(Nickolson_data['season'] == 'Spring') & (Nickolson_data['is_happy_customer'] == 1) & (Nickolson_data['is_expedited_delivery']==1)]
customer_Nickolson_data
```

### This Is a Nickolson Happy Customers And Orders data Also a spring Season selling

```python
plt.figure(figsize=(4, 4))
sns.boxplot(x='nearest_warehouse', y='order_price', data=customer_Nickolson_data)
plt.title('Nikolson Order Sales')
plt.show()
```

### We get Scatter Plot Of Nearest Warehouses

```python
plt.figure(figsize=(15, 15))
sns.pairplot(data=data, hue='nearest_warehouse')
plt.suptitle('Warehouse data', y=1.02)
plt.show()
```

```python
# Filter data for Thompson Warehouse
thompson_reviews = data[data['nearest_warehouse'] == 'Thompson']['is_happy_customer']

# Filter data for Nickolson Warehouse
nickolson_reviews = data[data['nearest_warehouse'] == 'Nickolson']['is_happy_customer']

# Plotting
plt.figure(figsize=(8, 6))
plt.hist([thompson_reviews, nickolson_reviews], bins=5, label=['Thompson', 'Nickolson'])
plt.xlabel('Customer Review')
plt.ylabel('Frequency')
plt.title('Customer Reviews by Warehouse')
plt.legend()
plt.show()
```

```python
# customers order in seasons


# Extracting 'season' and 'customer_id' columns
selected_data = data[['season', 'customer_id']]
selected_data
```

```python
#Storing the data in customers order

Customers_orders_warehouse=data[['season','customer_id']]
Customers_orders_warehouse
```

```python
#Storing the data in customers order

Customers_not_delivers_orders_warehouse=data[['nearest_warehouse','is_expedited_delivery']]
Customers_not_delivers_orders_warehouse
```

### in below Graph we get maximun sales in Spring Season

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='season', data=selected_data)
plt.title('Count of Orders by Season')
plt.xlabel('Season')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='nearest_warehouse', data=Customers_not_delivers_orders_warehouse)
plt.title('Count of not delivers Orders by Warehouse')
plt.xlabel('Warehouse')
plt.ylabel('Number of Orders')

plt.show()
```

### in below pie charts we get the most sales in seasons

```python
plt.figure(figsize=(6, 6))
data['season'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart Of Season ')
plt.ylabel('Seasons')
plt.show()
```

```python
# plotting pie chart 
plt.figure(figsize=(6, 6))
data['is_happy_customer'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie chart for Happy Customers ')
plt.ylabel('Happy Customers')
plt.show()
```
